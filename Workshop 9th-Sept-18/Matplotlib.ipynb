### Plots Using Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,7,9,10,16])
plt.plot(x)
plt.show()
```

```python
x = np.array([1,2,3,4,7,9,10,16])
plt.plot(x, x**2, label='2018')
plt.plot(x, x**3, label='2017')
plt.xlabel("No of days", color="white")
plt.ylabel("Prices", color="white")
plt.title("Stock Price Comparison", color="white")
plt.legend()
plt.show()
plt.plot(x, np.sin(x))
plt.show()
```

```python
y = np.array([1,2,3,4,5,6,7,8])
plt.scatter(x, y, color="Red", marker = "+")
plt.title("Stock Data", color="red")
plt.xlabel("Days")
plt.ylabel("Price")
plt.show()
```

```python
x = np.random.random((1000, 2))
for ix in range(x.shape[0]):
    plt.scatter(x[ix, 0], x[ix, 1], marker = '+', color = "Blue")
plt.show()
```

### Bar Chart

```python
numbers = np.random.randint(0, 10, 5)
indices = np.arange(0, 5)
numbers2 = np.random.randint(0, 10, 5)
indices2 = np.arange(0, 5)
print(numbers)
print(indices)
```

```python
plt.bar(indices, numbers, 0.25,  color='orange', label='Gold')
plt.bar(indices2+0.25, numbers2, 0.25,  color='green', label='Gold')
plt.legend()
plt.show()
```

### Pie Chart

```python
labels = ['Algebra', 'Coding', 'Algorithm', 'Calculus']
values = [90, 80, 50, 40]
plt.pie(values, labels=labels, radius=1, explode=(0, 0, 1, 0), shadow=True)
plt.show()
```

### Histogram

```python
# Generate a normal distribution(1D)
u = 5
sigma = 2
# For u = 0 & sigma = 1 Such a distribution is also called standard nomral distribution
vals = u + sigma*np.random.randn(1000)
plt.hist(vals, 50)
plt.show()
```

```python
x = vals
y = np.zeros(x.shape)
plt.scatter(x, y)
plt.show()
```

### Multivariate Normal Distribution

```python
mean = np.array([0.0, 0.0])
cov = np.array([[0.3, 0], [0, 1]])
dist = np.random.multivariate_normal(mean, cov, 500)
# plt.scatter(dist[:, 0], dist[:, 1], marker='+')
mean2 = np.array([5.0, 6.0])
cov2 = np.array([[1.3, 0.2], [0.2, 1.1]])
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
plt.scatter(dist2[:, 0], dist2[:, 1], color='red', marker='*')
```
